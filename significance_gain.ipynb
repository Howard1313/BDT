```python
import numpy as np  
import math
import matplotlib.pyplot as plt  

# s_base,b_base,db_base are signal, background and background error take in L=58fb^-1 cutflow in BDT cut
s_base = 2.63
b_base = 43.68
db_base = 5.99
x_list=[]
y_list=[]

def get_xy(formula, n):  
    x=0.1*n
    y = formula(x)  # <- note now we're calling the function 'formula' with x  
    x_list.append(x)
    y_list.append(y)


def my_formula(k):
    s = k*s_base
    b = k*b_base
    db= k*db_base
    sqrtInput = 2.*((s + b) * np.log( (s + b) * (b + db*db)/( b*b + (s + b) * db * db) )-b*b / (db*db) * np.log( 1. + db*db*s / (b * (b + db*db) ) ) )
    return math.sqrt(sqrtInput)
 
    
for n in range(10, 90):
    get_xy(my_formula, n)

plt.plot(x_list, y_list)  
plt.title("1u1e")
plt.xlabel("k (statistic gain)")
plt.ylabel('significance')
plt.xlim((0,10))
plt.ylim((0,2))
plt.show()

# 2.37,6.65 are statistic gain for L=58fb^-1 to L=139fb^-1 and L=58fb^-1 to L=389fb^-1
plt.plot(x_list, y_list)  
plt.title("1u1e")
plt.xlabel("k (statistic gain)")
plt.ylabel('significance')
plt.xticks([1,2.37,6.65])
plt.yticks([my_formula(1),my_formula(2.37),my_formula(6.65)])
plt.show()

print("sinificance gain (L=58->139):",my_formula(2.37)/my_formula(1))
print("sinificance gain (L=58->389):",my_formula(6.65)/my_formula(1))
```
