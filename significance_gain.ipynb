```python
import numpy as np  
import math
import matplotlib.pyplot as plt  

s_base = 2.63
b_base = 43.68
db_base = 5.99
x_list=[]
y_list=[]

def get_xy(formula, n):  
    x=0.1*n
    y = formula(s_base,b_base,db_base,x)  # <- note now we're calling the function 'formula' with x  
    x_list.append(x)
    y_list.append(y)

def my_formula(s_base,b_base,db_base,k):
    s = k*s_base
    b = k*b_base
    db= k*db_base
    sqrtInput = 2.*((s + b) * np.log( (s + b) * (b + db*db)/( b*b + (s + b) * db * db) )-b*b / (db*db) * np.log( 1. + db*db*s / (b * (b + db*db) ) ) )
    return math.sqrt(sqrtInput)

for n in range(10, 90):
    get_xy(my_formula, n)

plt.plot(x_list, y_list)  
plt.title("1u1e")
plt.xlabel("k (statistic gain)")
plt.ylabel('significance')
plt.xlim((0,10))
plt.ylim((0,2))
plt.show() 

```
